```python
def check_size(n=8):
    last = 0
    while True:
        plt.axis("off")
        plt.imshow(cv2.cvtColor(d[:,last:last+n], cv2.COLOR_BGR2RGB))
        plt.show()
        last += n
        if last > d.shape[1]:
            break
```

```python
check_size(10)
```

```python
check_size(5)
```

```python
def test_cut(n, m):
    plt.axis("off")
    plt.imshow(cv2.cvtColor(d[n:m,:], cv2.COLOR_BGR2RGB))
    plt.show()
    
```

```python
test_cut(n=0, m=160)
```
